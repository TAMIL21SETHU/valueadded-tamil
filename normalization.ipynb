```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import  Normalizer
```

```python
Normalization_df = pd.DataFrame({'x':np.random.randint(-100,100,1000),
                              'y':np.random.randint(-80,80,1000),
                              'z':np.random.randint(-150,150,1000)})
```

```python
Normalization_df
```

```python
from matplotlib.figure import projections
ax = plt.axes(projection='3d')

ax.scatter3D(Normalization_df.x,Normalization_df.y,
             Normalization_df.z,)

plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
nol= Normalizer()
```

```python
Normalization_df = nol.fit_transform(Normalization_df)
```

```python
Normalization_df
```

```python
Normalization_df=pd.DataFrame(Normalization_df,columns=['x','y','z'])
Normalization_df
```

```python
ax = plt.axes(projection='3d')

ax.scatter3D(Normalization_df.x,Normalization_df.y,
             Normalization_df.z,)

plt.figure()
```

```python
data={'first score':[100,90,np.nan,95],
      'second score':[30,45,56,np.nan],
      'third score':[np.nan,40,80,98]}
```

```python
data
```

```python
var=pd.DataFrame(data)
```

```python
var
```

```python
var.isnull()
var.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
imputer.fit(var)
imputer.transform(var)
```
